```python
def cat_experience(num):
    if num <= 2:
        return 'Fresher'
    elif 3 <= num <= 4:
        return 'Junior'
    elif 5 <= num <= 7:
        return 'Senior'
    elif 8 <= num <= 15:
        return 'Lead'
    elif 16 <= num <= 25:
        return 'Specialist'
    else:
        return 'Unknown'
```

```python
df['Years_of_Experience']= df['Years_of_Experience'].astype("int64")
```

```python
df['Category'] = df['Years_of_Experience'].apply(cat_experience)
```
